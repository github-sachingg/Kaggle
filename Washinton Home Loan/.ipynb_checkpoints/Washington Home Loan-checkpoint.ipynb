```python
#Loan originated : Means that the loan has been approved and lender has taken it/
# We can remove the below 
#Application withdrawn by applicant : Since this has nothing to do with nvestigation
#Loan purchased by the institution: These are loans which are purchased from secondary institutions 
#and hence nothing to do with the Applicant

loan_df = loan_df[loan_df['action_taken_name'] != 'Application withdrawn by applicant']
loan_df = loan_df[loan_df['action_taken_name'] != 'Loan purchased by the institution']

#loan_df = loan_df.drop([loan_df['action_taken_name'] == 'Loan purchased by the institution'],axis=0,inplace=True)
loan_df.shape
```

```python
#Lets read the County codes database
county_df = pd.read_excel(r"input\all-geocodes-v2016.xlsx")
```

```python
#Washington state code is 53
county_df.columns
```

```python
loan_df['state_name'].head()
```

```python
county_df = county_df[county_df['State Code (FIPS)'] == 53]
temp_df = pd.DataFrame()
```

```python
temp_df['county_name'] = county_df['Area Name (including legal/statistical area description)']
```

```python
temp_df['county_code'] = county_df['State Code (FIPS)'][county_df['State Code (FIPS)'] == 53]
```

```python
temp_df.head()
```

```python
temp_df.reset_index()
```

```python
temp_df.columns
#temp_df.drop('index',axis=1,inplace=True)
```

```python
temp_df.head()
```

```python
final_df = pd.merge(loan_df,temp_df,how='left',on='county_name')
```

```python
final_df.columns
```

```python
#Now we have final_df which is for Washington County
final_df['county_name'].unique()
```

```python
#Create a new feature 'Loan Approved'
final_df['Loan_Approved'] = final_df['action_taken_name'].map(lambda x: "Approved" if x=='Loan originated' else 'Not Approved')
```

```python
plt.figure(figsize=(20,20))
#sns.barplot(x='county_name',y='Loan_Approved',hue='applicant_sex_name',data=final_df)
final_df.plot(x='county_name',y='Loan_Approved',kind='bar')
```
