```python
train_df.isna().sum()
```

```python
def doSVM(X,Y,c=1):
    clf = SVC(kernel='rbf',C=c)
    clf.fit(X,Y)
    score = clf.score(X,Y)
    return clf, score
```

```python
Y_train = train_df.target
X_train = train_df
X_train.drop('target',axis=1,inplace=True)
X_train.drop('id',axis=1,inplace=True)
```

```python
from sklearn.metrics import mean_squared_error
C = [100,10,1,0.1,0.01,0.001]
output = []
m,n = X_train.shape

for cval in C:
    model,score = doSVM(X_train,Y_train,cval)
    pred = model.predict(X_train)
    meanErr = mean_squared_error(Y_train,pred)
    #meanErr = sum(((Y_train - pred)))
    output.append((cval,score,meanErr))

output_df = pd.DataFrame(output,columns=['C','Score','Error'])
output_df
```

```python
test_df = pd.read_csv(r"C:\Users\sa952592\Documents\Python Scripts\Kaggle\Dont Overfit\input\test.csv")
id_df = test_df.id
```
